```python
# ============================================================
# Project: Karg & Thompson Algorithm - Closed Jobs Sequence Route Minimizer
# Author:  Pier Sergio Caltabiano
# Date:    March 06, 2026
# Description: Implementation of a heuristic insertion algorithm
#              to find the local optimal sequence of jobs minimizing
#              the total travel time across all tasks.
# ============================================================

import random # Import the random module to enable random job selection
```

```python
# STEP 1: Job Input 

jobs=[] # List to store job names
job_count=1 # Counter to track the number of jobs entered

print("Enter job names (maximum 10). Type 'done' to finish.\n")

while job_count <= 10:
    name = input(f"Enter the name of job #{job_count}: ").strip()  # Remove leading/trailing spaces
    
     # If user wants to stop early
        
    if name.lower() == "done":
        if len(jobs) < 2:
            print("Please enter at least 2 jobs before finishing.")
            continue
        break
    
    # Reject empty names
    
    if not name:
        print("The name cannot be empty.")
        continue
    
    jobs.append(name) # Add valid job name to the list
    job_count += 1

# Notify user if maximum limit is reached    
    
if len(jobs) == 10:
    print("You have reached the maximum of 10 jobs.")

print("\nJobs entered:", jobs)
```

```python
# STEP 2: Travel Time Matrix 

n = len(jobs) # Number of jobs

# Initialize an n x n matrix filled with None (diagonal will remain None)

travel_matrix = [[None] * n for _ in range(n)] 

# Populate the matrix by asking the user for travel times between each pair of jobs

for i in range(n):                
    for j in range(n):                              
        if i != j: # Skip diagonal (a job cannot travel to itself)                                        
            travel_time = input(f"Time from {jobs[i]} to {jobs[j]}: ")                                         
            travel_matrix[i][j] = int(travel_time)  # Convert string input to integer
```

```python
# STEP 3: Cost Calculation 

def calc_cost(sequence):
    
    #Calculates the total travel time for a given sequence of jobs.
    #The route is closed, meaning it returns to the first job at the end.
        
    route_cost = 0
    
    # Sum travel times between each consecutive pair of jobs
    for i in range(len(sequence) - 1):
        route_cost  = route_cost  + travel_matrix[jobs.index(sequence[i])][jobs.index(sequence[i+1])]
    
    # Add the return trip from the last job back to the first (closed route)
    route_cost  = route_cost  + travel_matrix[jobs.index(sequence[-1])][jobs.index(sequence[0])]
    
    return route_cost 
```

```python
# STEP 4: Heuristic Insertion Algorithm 

# Start with a random pair of jobs as the initial sequence
sequence = random.sample(jobs, 2)

# Build a list of jobs not yet included in the sequence
remaining = [job for job in jobs if job not in sequence]

# Insert each remaining job into the best possible position
for job in remaining:
    best_cost = float('inf') # Initialize with infinity so any real cost will be lower
    best_position = None # Will store the index of the best insertion point
    
    # Try inserting the job at every valid position (never before the first job)
    for i in range(1, len(sequence) + 1):
        copy = sequence.copy() # Work on a copy to avoid modifying the original
        copy.insert(i, job) # Insert job at position i
        cost = calc_cost(copy) # Calculate the total route cost
        if cost < best_cost:
            best_cost = cost # Update best cost found so far
            best_position = i # Update best position found so far
    
    sequence.insert(best_position, job) # Insert job at the optimal position

    
# OUTPUT

print("Optimal Sequence*:", sequence)
print("Total Time:", calc_cost(sequence))
print(f"""
*Note: The sequence shown above is generated using a heuristic insertion algorithm.
The method builds the route step by step by inserting each task in the position that
minimizes the current cycle cost, starting from a random initial pair of tasks.

Because the algorithm makes locally optimal decisions rather than evaluating all
possible permutations, the resulting sequence is not guaranteed to be globally optimal.
Different runs may therefore produce different sequences and total costs.

In some cases, the structure of the heuristic may prevent certain sequences from ever
being generated, meaning the true global optimum might never be found even after many
executions. The reported result should therefore be interpreted as a good feasible
solution, but not necessarily the true optimal one.
""")
```
